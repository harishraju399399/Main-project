```python
HR_Employee_Attrition['Attrition'].value_counts(normalize=True)
```

```python
##department wise attrition ##

Depatmentwise_attrition=HR_Employee_Attrition[HR_Employee_Attrition['Attrition']=='Yes']['Department'].value_counts(normalize=True)
```

```python
print(Depatmentwise_attrition)
```

```python
##educationwise attrition##
Educationwise_attrition=HR_Employee_Attrition[HR_Employee_Attrition['Attrition']=='Yes']['EducationField'].value_counts(normalize=True)
```

```python
print(Educationwise_attrition)
```

```python


##Agewise attrition #


Agewise_attrition=HR_Employee_Attrition[HR_Employee_Attrition['Attrition']=='Yes']['Age'].value_counts(normalize=True)



```

```python
print(Agewise_attrition)
```

```python
#Jobrole attrition##

Jobrolewise_attrition=HR_Employee_Attrition[HR_Employee_Attrition['Attrition']=='Yes']['JobRole'].value_counts(normalize=True)
```

```python
print(Jobrolewise_attrition)
```

```python
sns.barplot(x=Depatmentwise_attrition.index, y=Depatmentwise_attrition.values, palette='Blues')
plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```
